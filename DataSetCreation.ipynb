```python
from mtcnn import MTCNN
import cv2
import os
```

```python
folder=r"C:\Users\abhin\Desktop\ml\Facial Rrecognition\Dataset\Abhinav"
Abhinav=[]
for filename in os.listdir(folder):
    img=cv2.imread(os.path.join(folder,filename))

    Abhinav.append([0,img])
        
        
    
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(Abhinav[10][1])
```

```python

detector =MTCNN()

```

```python
face=detector.detect_faces(Abhinav[1][1])
```

```python
bx=face[0]["box"]
fc=Abhinav[1][1][bx[1]:bx[1]+bx[3],bx[0]:bx[0]+bx[2]]
plt.imshow(fc)
print(face[0]["confidence"])
    
```

```python
Images=[]
for i in range(len(Abhinav)):
    face=detector.detect_faces(Abhinav[i][1])
    for y in face:
        if(y["confidence"]>0.90):
           bx=y["box"]
           Images.append(Abhinav[i][1][bx[1]:bx[1]+bx[3],bx[0]:bx[0]+bx[2]]) 
        
    
```
